### 1.3 Get latitude and longitude for those neighboods.

```python
def get_neighborhood_coordinate(neighborhood):
    lat_lng_coords = None
    # loop until you get the coordinates
    while(lat_lng_coords is None):
      g = geocoder.arcgis('{}, Raleigh, NC'.format(neighborhood))
      lat_lng_coords = g.latlng

    return lat_lng_coords
```

```python
get_neighborhood_coordinate("West Raleigh Historic District")
```

```python
coords = [get_neighborhood_coordinate(name) for name in raleigh_neighborhoods['neighborhood_name']]
```

```python
df_coords = pd.DataFrame(coords, columns=['latitude', 'longitude'])
df_coords.head()
```

```python
raleigh_neighborhoods['latitude'] = df_coords['latitude']
raleigh_neighborhoods['longitude'] = df_coords['longitude']
raleigh_neighborhoods.head()
```

### 2.3 Draw Raleigh NC Map

```python
# get the coordinates of Kuala Lumpur
address = 'Raleigh, NC, USA'

geolocator = Nominatim(user_agent="my-application")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Raleigh, NC are: {}, {}.'.format(latitude, longitude))
```

```python
# draw raleigh map:
map_raleigh = folium.Map(location=[latitude, longitude], zoom_start=10)
```

```python
type(raleigh_neighborhoods.head())
```

```python
# add markers to map
df = raleigh_neighborhoods
print(df.shape)
for lat, lng, neighborhood in zip(df['latitude'], df['longitude'], df['neighborhood_name']):
    label = '{}, {}'.format(neighborhood, 'Raleigh')
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7).add_to(map_raleigh)
map_raleigh
```

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
        #print(url)    
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'neighborhood_Latitude', 
                  'nighborhood_Longitude', 
                  'venue', 
                  'venue_Latitude', 
                  'venue_Longitude', 
                  'venue_Category']
    return(nearby_venues)
```

```python
def getNearbyCateId(cateId, names, latitudes, longitudes, radius=500):
    venues_list=[]
    LIMIT = 2000
    #cateId = '4bf58dd8d48988d17f941735'   # movie theator
    for name, lat, lng in zip(names, latitudes, longitudes):
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}&categoryId={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT,
            cateId)
        #print(url)    
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['neighborhood_name', 
                  'neighborhood_latitude', 
                  'nighborhood_longitude', 
                  'venue', 
                  'venue_latitude', 
                  'venue_longitude', 
                  'venue_category']
    return(nearby_venues)
```
